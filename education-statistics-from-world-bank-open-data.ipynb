```python

years = df2.columns[4:]
summary = pd.DataFrame({
    'Year': years,
    'Skewness': [df2[year].skew() for year in years],
    'Kurtosis': [df2[year].kurtosis() for year in years]
})

print(summary.head())

```

```python
mean_skewness = summary['Skewness'].mean()
mean_kurtosis = summary['Kurtosis'].mean()

print(f"Mean Skewness: {mean_skewness}")
print(f"Mean Kurtosis: {mean_kurtosis}")
```

The median is generally a safer choice regardless of the distribution shape because it is less affected by outliers. To understand the data distribution in detail, I analyzed the skewness and kurtosis values for each year individually. This helped identify any deviations from normality and provided insights into the nature of the data for each column.

After calculating skewness and kurtosis year by year, I summarized the results to make the evaluation more manageable. The average skewness value across all years was found to be 65.3333, which is extremely high and positive, indicating a highly right-skewed distribution. This means that a large portion of the data is clustered on the left side of the distribution, with very few data points on the right side. Similarly, the average kurtosis value was 5624.0505, which is also extremely high, suggesting the presence of many extreme values. This indicates that the distribution is very peaked and has numerous outliers.

Both the high positive skewness and high kurtosis confirm that the data significantly deviates from a normal distribution. Most of the data is skewed to the left, with a large number of outliers on the right side. Given these characteristics, filling missing values with the mean would not be appropriate. Instead, the median was chosen as a fill strategy, as it is less sensitive to outliers and better suited for right-skewed data with extreme values.ues.

```python
#To fill missing values in all columns except the 'object' type columns with the median.
df2 = df2.apply(lambda x: x.fillna(x.median()) if x.dtype != 'object' else x)
df2.info()
```

```python
for year in range(1970, 2018):  
    latest_year = str(year)  
    top_countries = df2.groupby('Country Name')[latest_year].max().nlargest(5).index
    bottom_countries = df2.groupby('Country Name')[latest_year].min().nsmallest(5).index
    print(f"Top Countries in {latest_year}:", top_countries.tolist())
    print(f"Bottom Countries in {latest_year}:", bottom_countries.tolist())
    print("-" * 50)
```
